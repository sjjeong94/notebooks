```python
import torch
import numpy as np
```

```python
x0 = np.random.randn(1000000).astype(np.float32)
x1 = x0.copy()

np.random.shuffle(x1)

print('%.12f' % (x0.sum() - x1.sum()))
print('%.12f' % (np.sort(x0).sum() - np.sort(x1).sum()))
```

```python
x0 = np.random.randn(1000000).astype(np.float32)
x1 = x0[::-1].copy()

print('%.12f' % (x0.sum() - x1.sum()))
print('%.12f' % (np.sort(x0).sum() - np.sort(x1).sum()))
```

```python
x0 = torch.randn(1000000)
x1 = x0.cuda()

s0 = x0.sum()
s1 = x1.sum().cpu()

print('%.12f' % (s0 - s1))
```
